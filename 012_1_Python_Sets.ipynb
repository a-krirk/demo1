#เซต Python Sets

เซต (Set)ชุดใช้เพื่อเก็บหลายรายการในตัวแปรเดียว

Set เป็นหนึ่งใน 4 ประเภทข้อมูลในตัว(built-in)ใน Python ที่ใช้เพื่อจัดเก็บคอลเลกชั่นข้อมูล อีก 3 ประเภทคือ List, Tuple และ Dictionary ซึ่งทั้งหมดมีคุณสมบัติและการใช้งานที่แตกต่างกัน

เซต คือคอลเล็กชันที่ไม่มีลำดับ(unordered)และไม่มีดัชนี(unindexed)

เซต เขียนด้วยวงเล็บปีกกา

```python
thisset = {"mango", "banana", "coconut"}
print(thisset)
```

##รายการเซต(Set Items)

รายการเซต ไม่มีลำดับ(unordered) ไม่สามารถเปลี่ยนแปลงได้(unchangeable) และไม่อนุญาตให้มีค่าที่ซ้ำกัน(duplicate values not allow)



##ไม่มีลำดับ(Unordered) 

ไม่มีลำดับมายความว่ารายการในชุดไม่มีลำดับที่กำหนดไว้

รายการเซตสามารถปรากฏในลำดับที่แตกต่างกันทุกครั้งที่คุณใช้งาน และไม่สามารถอ้างอิงด้วยดัชนีหรือคีย์ได้


#ไม่สามารถเปลี่ยนแปลงได้(Unchangeable)

เซต ไม่สามารถเปลี่ยนแปลงได้ หมายความว่าเราไม่สามารถเปลี่ยนรายการหลังจากสร้างชุดแล้ว



##ไม่อนุญาตให้มีค่าที่ซ้ำกัน(Duplicate values not allow)

เซตต้องไม่มีค่าเท่ากันสองรายการ


```python
thisset = {"apple", "banana", "coconut", "apple"}

print(thisset)
```

##การหาค่าความยาวของเซต (Get the Length of a Set)

เพื่อหาจำนวนรายการที่มีในเซต สามารถใช้ len()


```python
thisset = {"apple", "coconut", "mango"}

print(len(thisset))
```

##รายการเซต ประเภทข้อมูล (Set Items - Data Types)

รายการเซต สามารถเป็นข้อมูลประเภทใดก็ได้



```python
set1 = {"apple", "mango", "coconut"}
set2 = {1, 5, 8, 10, 3}
set3 = {True, False, False}
```

เซต สามารถประกอบด้วยข้อมูลหลายประเภท

```python
set1 = {"นาย สมศักดิ์ ศักดิ์สม", 34, True, 40, "male"}
```

#ประเภท type()

เซตถูกกำหนดเป็นวัตถุที่มีประเภทข้อมูล 'set' จากมุมมองของPython



```python
myset = {"apple", "mango", "coconut"}
print(type(myset))
```

##ตัวสร้างเซต [set() Constructor]

สามารถใช้ set() เพื่อสร้างเซต
It is also possible to use the set() constructor to make a set.

```python
thisset = set(("apple", "mango", "coconut")) # สังเกตุการใช้วงเล็บ2อันซ้อนกัน
print(thisset)
```

##คอลเลกชัน Python (อาร์เรย์) [Python Collections] (Arrays)]

ในภาษาการเขียนโปรแกรม Python มีคอเล็กชั่นของข้อมูล 4 ประเภท
* รายการ(List)เป็นคอลเลกชันที่สั่งและเปลี่ยนแปลงได้ อนุญาตให้สมาชิกที่ซ้ำกัน
* ทูเพิล(Tuple)คือคอลเล็กชันที่สั่งซื้อและไม่สามารถเปลี่ยนแปลงได้ อนุญาตให้สมาชิกที่ซ้ำกัน
* เซ็ท(set)คือคอลเล็กชันที่ไม่มีลำดับและไม่ได้จัดทำดัชนี ไม่มีสมาชิกซ้ำ
* พจนานุกรม(dictionary)คือคอลเล็กชันที่เรียงลำดับได้ เปลี่ยนแปลงได้ ไม่มีสมาชิกซ้ำ
