# Python - Nested Dictionaries

## Nested Dictionaries
dictionary ที่มี dictionary อื่นซ้อนอยู่ด้านใน เรียกว่า nested dictionaries

### ตัวอย่าง
สร้าง dictionary ซ้อน dictionary

```python
myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}
```

### ตัวอย่าง
สร้าง dictionary 1 อัน ที่มี dictionary 3 อันย่อยซ้อนอยู่ด้านใน

```python
child1 = {
  "name" : "Emil",
  "year" : 2004
}
child2 = {
  "name" : "Tobias",
  "year" : 2007
}
child3 = {
  "name" : "Linus",
  "year" : 2011
}

myfamily = {
  "child1" : child1,
  "child2" : child2,
  "child3" : child3
}
```
