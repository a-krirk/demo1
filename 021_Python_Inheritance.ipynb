#Python Inheritance

##Python Inheritance
Inheritance คือการอนุญาตให้ส่งต่อคุณสมบัติและวิธีการทั้งหมดของ class นึงให้กับ class อื่นๆ

Parent class คือ class ที่ทำหน้าที่ส่งต่อคุณสมบัติ เรียกว่า base class

Child class คือ class ที่รับคุณสมบัติจาก class อื่นๆ เรียกว่า derived class


##สร้าง Parent Class

class ใดๆก็สามารถที่จะเป็น parent class ได้ ดังนั้น ไวยากรณ์ที่ใช้ในการสร้างจึงเหมือนกับการสร้าง class ทั่วๆไป


##ตัวอย่าง

สร้าง class ชื่อว่า Food ซึ่งประกอบด้วยตัวแปร menu, price และ ฟังก์ชั่น printfood




```python
class Food:
  def __init__(self, menu, nationality):
    self.menufood = menu
    self.nationalityfood = nationality

  def printfood(self):
    print(self.menufood, self.nationalityfood)

#Use the Food class to create an object, and then execute the printfood method:

x = Food("Sushi", "Japanese")
x.printfood()
```

#สร้าง Child Class

การสร้าง class เพื่อรับคุณสมบัติจาก class อื่นๆ เราจะกำหนด parameter ของ child class เป็นค่าจาก parent class:



##ตัวอย่าง

สร้าง class ชื่อว่า Meal และรับคุณสมบัติกับวิธีการจาก class Food:



```python
class Meal(Food):
  pass
```

ตอนนี้ class Meal มีคุณสมบัติและวิธีเหมือนกับ class Food 




##ตัวอย่าง

ใช้ class Meal ในการสร้าง Object จากนั้นแสดงค่า printfood :





```python
x = Meal("Sushi", "Japanese")
x.printfood()
```

##เพิ่มฟังก์ชั่น __init__()

ตอนนี้เราได้สร้าง child class เพื่อถ่ายทอดคุณสมบัติและวิธีการจาก parent class

เราต้องการที่จะเพิ่มฟังก์ชั่น __init__() ให้กับ child class ( แทนที่ keyword pass )



##ตัวอย่าง

เพิ่มฟังก์ชั่น __init__() ใน Meal class :


```python
class Meal(Food):
  def __init__(self, menu, nationality):
    #add properties etc.
```

เมื่อเราเพิ่มฟังก์ชั่น __init__() child class จะไม่ส่งต่อคุณสมบัติจากฟังชั่น __init__() ของ class parent




การรักษาคุณสมบัติการส่งต่อของฟังก์ชั่น __init__() ใน parent class สามารถทำได้โดยการเรียกฟังก์ฟังก์ชั่น __init__() ของ parent class





##ตัวอย่าง


```python
class Meal(Food):
  def __init__(self, menu, nationality):
    Food.__init__(self, menu, nationality)
```

ตอนนี้เราสามารถเพิ่มฟังก์ชั่น __init__() ได้สำเร็จ และเก็บคุณสมบัติการส่งต่อของ parent class และเราพร้อมที่จะเพิ่ม functionality ในฟังก์ชั่น __init__()




##ใช้ฟังก์ชั่น super()

Python ยังมีฟังก์ชั่น super() ที่จะส่งต่อคุณสมบัติและวิธีการทั้งหมดของ class parent ไปยัง class child :




```python
class Meal(Food):
  def __init__(self, menu, nationality):
    super().__init__(menu, nationality)
```

การใช้ฟังก์ชั่น super() เราไม่จำเป็นที่จะต้องใช้ชื่อของ parent element เนื่องจาก มันจะส่งต่อคุณสมบัติและวิธีการจาก parent class โดยอัตโนมัติ
