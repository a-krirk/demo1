#Python Try Except

block ของคำสั่ง **"try"** ใช้ตรวจจับ error ของโค้ด

block ของคำสั่ง **"except"** ใช้จัดการกับ error นั้น

block สุดท้ายทำให้รันโค้ดได้ โดยผลลัพธ์ไม่เกี่ยวข้อง block try และ except

##Exception Handling


เมื่อมีข้อผิดพลาดเกิดขึ้น ปกติ Python จะหยุดรันแล้วส่งข้อความ error กลับมา

ข้อผิดพลาดต่าง ๆ แก้ได้ด้วยการใช้คำสั่ง **try**

###ตัวอย่าง
block **try** จะเกิด error ขึ้น เพราะไม่ได้ระบุค่าของ x

```python
try:
  print(x)
except:
  print("An exception occurred")
```

เมื่อ block **try** เกิด error ขึ้น block **except** ก็จะรันต่อ
Since the try block raises an error, the except block will be executed.

หากไม่มี block **try** โปรแกรมจะหยุดทำงานและแสดงข้อความ error ออกมา

```python
print(x)
```

##Many Exceptions

สามารถสร้าง exception ได้หลาย ๆ block เช่น ถ้าต้องการจะรันโค้ดใน block ที่ต้องการของ error บางประเภท

###ตัวอย่าง
Print one message if the try block raises a NameError and another for other errors:

```python
try:
  print(x)
except NameError:
  print("Variable x is not defined")
except:
  print("Something else went wrong")
```

##Else

You can use the else keyword to define a block of code to be executed if no errors were raised:

###ตัวอย่าง


In this example, the try block does not generate any error:

```python
try:
  print("Hello")
except:
  print("Something went wrong")
else:
  print("Nothing went wrong")
```

##Finally

The finally block, if specified, will be executed regardless if the try block raises an error or not.

###ตัวอย่าง


```python
try:
  print(x)
except:
  print("Something went wrong")
finally:
  print("The 'try except' is finished")
```

This can be useful to close objects and clean up resources:

###ตัวอย่าง
Try to open and write to a file that is not writable:


```python
try:
  f = open("demofile.txt")
  try:
    f.write("Lorum Ipsum")
  except:
    print("Something went wrong when writing to the file")
  finally:
    f.close()
except:
  print("Something went wrong when opening the file")
```

The program can continue, without leaving the file object open.

##Raise an expection

As a Python developer you can choose to throw an exception if a condition occurs.

To throw (or raise) an exception, use the raise keyword.

###ตัวอย่าง


Raise an error and stop the program if x is lower than 0:

```python
x = -1

if x < 0:
  raise Exception("Sorry, no numbers below zero")
```

The raise keyword is used to raise an exception.

You can define what kind of error to raise, and the text to print to the user.

###ตัวอย่าง
Raise a TypeError if x is not an integer:

```python
x = "hello"

if not type(x) is int:
  raise TypeError("Only integers are allowed")
```
