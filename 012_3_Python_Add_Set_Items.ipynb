## เพิ่ม Sets
การเพิ่มข้อมูล set นึงเข้าในอีก set นึงเราจะใช้คำสั่ง update()


## ตัวอย่าง
เพิ่มข้อมูลจาก meal ไปยัง foodset:

```python
foodset = {"Sushi", "Burger", "Salad"}
meal = {"Pizza", "Noodle", "Curry"}

foodset.update(meal)

print(foodset)
```

## เพิ่ม Iterable ใดๆ
object ที่ใช้กำบคำสั่ง update() ไม่จำเป็นที่จะต้องเป็น set เท่านั้น สามารถเป็น iterable ใดๆก็ได้ (tuples, lists, dictionaries, ฯลฯ)


## ตัวอย่าง
เพิ่มข้อมูลจาก list เข้าใน set :


```python
foodset = {"Sushi", "Burger", "Salad"}
mylist = ["Pizza", "Noodle"]

foodset.update(mylist)

print(foodset)
```
