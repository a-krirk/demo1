#Python Dates

วันที่ใน Python ไม่ได้จัดเป็นประเภทข้อมูล แต่เราสามารถ import module ชื่อ **datetime** เพื่อมาทำงานร่วมกับวัน-เวลา

```python
import datetime

x = datetime.datetime.now()
print(x)
```

##Date Output

เมื่อเรารันโค้ดจากตัวอย่างด้านบน จะได้ผลลัพธ์คือ:
2021-08-31 15:29:38.776080

วันที่จะประกอบไปด้วยปี เดือน วัน ชั่วโมง นาที วินาที และมิลลิวินาที

ใน module **datetime** มีหลายวิธีที่จะ return ค่าเกี่ยวกับวัน-เวลาออกมา ซึ่งจะมีตัวอย่างให้ได้เรียนรู้ในบทต่อ ๆ ไป

```python
import datetime

x = datetime.datetime.now()

print(x.year)
print(x.strftime("%A"))
```

function  **abs()** จะให้ผลลัพธ์เป็น absolute (ค่าบวก) ของเลขนั้น ๆ

```python
x = abs(-8.5)

print(x)
```

function pow(x, y)  จะให้ผลลัพธ์ของ x ยกกำลัง y (x^y)

```python
x = pow(5, 2)

print(x)
```

##การสร้าง Date Objects

ใช้ class **datetime()** ใน module **datetime** ในการสร้างวัน-เวลา 

class **datetime()** ต้องการ 3  parameters คือ ปี, เดือน, ปี

```python
import datetime

x = datetime.datetime(2020, 5, 17)

print(x)
```

class **datetime()** สามารถใส่ parameter สำหรับ time & timezone (ชั่วโมง, นาที, วินาที, มิลลิวินาที, timezone) โดยมีค่า default เป็น 0 และค่า None สำหรับ timezone

##Method strftime()



object **datetime** มี method สำหรับปรับเปลี่ยนรูปแบบของวันที่เป็น string


method นี้คือ **strftime()** โดยจะรับค่ามา 1 parameter เพื่อระบุ format ที่ต้องการให้ return ค่าออกมาเป็น string

```python
import datetime

x = datetime.datetime(2018, 6, 1)

print(x.strftime("%B"))
```

Parameter ของสำคัญของ function **strftime()**

* "%a" = ชื่อของวันในสัปดาห์ (แบบย่อ) เช่น Sun - Sat
* "%A" = ชื่อของวันในสัปดาห์ (แบบเต็ม) เช่น Sunday - Saturday
* "%w" = หมายเลขของวันในสัปดาห์ ตั้งแต่ 0 - 6 เช่น 0 คือวันอาทิตย์, 6 คือวันเสาร์
* "%d" = วันของเดือน โดยมี 0 นำหน้า (01-31)
* "%b" = ชื่อของเดือน (แบบย่อ) เช่น Jan - Dec
* "%B" = ชื่อของเดือน  (แบบเต็ม) เช่น January - December
* "%m" = หมายเลขเดือนของปี  โดยมี 0 นำหน้า (01-12) เช่น 01 คือวันที่ 1
* "%y" = ปีแบบย่อ มีตัวเลขสองตัว (00 - 99) 
* "%Y" = ปีแบบย่อ มีตัวเลขสี่ตัว (0000 - 9999)
* "%H" = ชั่วโมงในรูปแบบ 24 ชั่วโมง (00-23)
* "%I" = ชั่วโมงในรูปแบบ 12 ชั่วโมง (00-12)
* "%p" = ตัวบอกเวลาก่อนและหลังเที่ยง (AM/PM)
* "%M" = นาทีในรูปแบบสองตัวเลข (00-59)
* "%S" = วินาทีในรูปแบบสองตัวเลข (00-59)
* "%f" = ไมโครวินาทีในรูปแบบหกตัวเลข (000000-999999)
* "%z" = เขตเวลา UTC ในรูปแบบ +HHMM เช่น +0000, +0730, -1200
* "%Z" = ชื่อเขตเวลา เช่น UTC, GMT
* "%j" = วันของปีมี 0 นำหน้า (001-366)
* "%U" = สัปดาห์ของปีมี 0 นำหน้า โดยมีวันอาทิตย์เป็นวันแรกของสัปดาห์ (00 - 53)
* "%W" = สัปดาห์ของปีมี 0 นำหน้า โดยมีวันจันทร์เป็นวันแรกของสัปดาห์ (00 - 53)
* "%c" = วันที่และเวลาในท้องถิ่นปัจจุบัน
* "%C" = ศตวรรษ
* "%x" = วันที่ในท้องถิ่นปัจจุบัน เช่น 09/27/2021
* "%X" = เวลาในท้องถิ่นปัจจุบัน เช่น 15:30:00
* "%%" = ใช้สำหรับแสดงตัวอักษร
* "%G" = ปีในรูปแบบสี่ตัวเลข เช่น 2021
* "%u" = วันในสัปดาห์ (1-7) เช่น 1
* "%V" = สัปดาห์ในปี (01-53) เช่น 01
