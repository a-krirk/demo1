## แบบฝึกหัด
ตรวจสอบว่า "Sushi" อยู่ใน set food หรือไม่

```python
food = {"Sushi", "Burger", "Salad"}
# your code here, replace __ with your answer.
if "Sushi" __ food:
  print("Yes, sushi is a food!")
```
