#Python Classes and Objects



##Python Classes/Objects


Python เป็น Object ที่มุ่งเน้นเรื่องภาษาโปรแกรม

เกือบทุกอย่างใน Python เป็น Object ที่มีคุณสมบัติและวิธีการ

Class เป็นเหมือนส่วนประกอบของ Object หรือ พิมพ์เขียวในการสร้าง Object

-----------------------------------------------------------------------

Python is an object oriented programming language.

Almost everything in Python is an object, with its properties and methods.

A Class is like an object constructor, or a "blueprint" for creating objects.

##สร้าง Class
การสร้าง Class เราจะใช้คำสั่ง "class"




##ตัวอย่าง

สร้าง class ชื่อว่า MyClass ที่มีตัวแปรชื่อ x




```python
class MyClass:
  x = 10
```

##สร้าง Object

ตอนนี้เราสามารถใช้ class ที่ชื่อว่า MyClass ในการสร้าง Objects :




##ตัวอย่าง

สร้าง object ชื่อว่า b1 และแสดงค่าของ x



```python
b1 = MyClass()
print(b1.x)
```

##ฟังก์ชั่น init()

จากตัวอย่างข้างต้นจะเป็นรูปแบบง่ายๆของ class และ object ซึ่งยังไม่สามารถนำไปใช้ประโยชน์ได้ในชีวิตจริง

เพื่อให้เข้าใจความหมายของ class เราต้องเข้าใจกับฟังก์ชั่น __init__() กันก่อน

class ทั้งหมดเป็นฟังก์ชั่นที่เรียกว่า  __init__() ซึ่งจะถูกใช้เสมอเมื่อเริ่มใช้ class

เราจะใช้ฟังก์ชั่น __init__() เพื่อกำหนดคุณสมบัติให้กับ object หรือ operation อื่นๆที่จำเป็นเมื่อสร้าง object :

--------------------------------------------------------------------------
## <font color="lightblue"> The __init__() Function </font>
The examples above are classes and objects in their simplest form, and are not really useful in real life applications.

To understand the meaning of classes we have to understand the built-in __init__() function.

All classes have a function called __init__(), which is always executed when the class is being initiated.

Use the __init__() function to assign values to object properties, or other operations that are necessary to do when the object is being created:

##ตัวอย่าง
สร้าง class ชื่อว่า Food และใช้ฟังก์ชั่น __init__() ในการกำหนดค่าตัวแปร menu และ price :

----------------------------------------------------------------------------

## <font color="lightblue"> Example </font>
Create a class named Person, use the __init__() function to assign values for name and age:



```python
class Food:
  def __init__(self, menu, price):
    self.menu = menu
    self.price = price

b1 = Food("burger", 50)

print(b1.menu)
print(b1.price)
```

## Object Methods

เรายังสามารถที่จะกำหนด method ใน Object ได้ โดย method ใน object จะเป็นฟังก์ชั่นของ object 

จงสร้าง method ใน class Food :

--------------------------------------------------------------------------------

## <font color="lightblue"> Object Methods </font>
Objects can also contain methods. Methods in objects are functions that belong to the object.

Let us create a method in the Person class:

##ตัวอย่าง

เพิ่มฟังก์ชั่นแสดงคำถามพร้อมกับ object b1





```python
class Food:
  def __init__(self, menu, price):
    self.menu = menu
    self.price = price

  def myfunc(self):
    print("คุณชอบกิน" + self.menu + "มั้ย?")

b1 = Food("Burger", 50)
b1.myfunc()
```

## The self Parameter
self parameter จะอ้างอิงไปยัง class ปัจจุบัน และใช้เพื่อเข้าถึงตัวแปรใน class

เราสามารถที่จะตั้งชื่อได้ตามที่ต้องการ ไม่จำเป็นที่จะต้องตั้งชื่อว่า self แต่ parameter self จะต้องเป็นตัวแรกในฟังก์ชั่นนั้นๆ ใน class :

##ตัวอย่าง
เราใช้คำว่า myfood และ botnoi แทนคำว่า self


```python
class Food:
  def __init__(myfood, menu, price):
    myfood.menu = menu
    myfood.price = price

  def myfunc(botnoi):
    print("บอทน้อยชอบกิน" + botnoi.menu)

b1 = Food("ฺBurger", 50)
b1.myfunc()
```

##แก้ไขคุณสมบัติของ Object

เราสามารถแก้ไขคุณสมบัติของ Object ได้ดังนี้ :


##ตัวอย่าง
กำหนดให้ price ของ b1 เท่ากับ 60




```python
b1.price = 60
```

##ลบคุณสมบัติของ Object
เราสามารถลบคุณสมบัติของ Object โดยใช้คำสั่ง del


##ตัวอย่าง

ลบคุณสมบัติของ price จาก Object b1 :


```python
del b1.price
```

##ลบ Object

เราสามารถลบ Object โดยใช้คำสั่ง del





##ตัวอย่าง
