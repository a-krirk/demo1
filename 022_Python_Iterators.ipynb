#Python Iterators


##Python Iterators

iterator คือ Object ที่ประกอบด้วยตัวแปรที่นับได้

iterator คือ Object ที่สามารถวนซ้ำได้ หมายความว่าเราสามารถที่จะวนเข้าถึงทุกๆค่าได้

ในทางเทคนิคของ python iterator เป็น object ที่ใช้ protocal วนซ้ำ ซึ่งประกอบด้วย method __iter__() และ __next__()



##Iterator เทียบกับ Iterable

Lists, tuples, dictionaries, และ sets ทั้งหมดเป็น iterable objects และเป็น iterable container ที่มาจาก iterator

ทั้งหมดเป็น Object ที่ได้มาจากการวนซ้ำโดยใช้ iter():

##ตัวอย่าง

ส่งคืนค่า iterator ที่ได้จาก tuple และแสดงค่าแต่ละตัว:



```python
mytuple = ("Burger", "Sushi", "Salad")
myfood = iter(mytuple)

print(next(myfood))
print(next(myfood))
print(next(myfood))
```

แม้แต่ string ก็สามารถที่จะวนซ้ำแล้วส่งคืนค่าออกมาได้ :



##ตัวอย่าง

String มีลำดับของตัวอักษร จึงเป็น object ที่สามารถวนซ้ำได้ :



```python
mystr = "botnoi"
mychar = iter(mystr)

print(next(mychar))
print(next(mychar))
print(next(mychar))
print(next(mychar))
print(next(mychar))
print(next(mychar))
```

##การวนซ้ำด้วย Iterator

เราสามารถใช้ for loop ในการวนซ้ำได้ :


##ตัวอย่าง

วนซ้ำค่าของ tuple :


```python
mytuple = ("Burger", "Sushi", "Salad")

for x in mytuple:
  print(x)
```

##ตัวอย่าง

วนซ้ำตัวอักษรของ string :


```python
mystr = "botnoi"

for x in mystr:
  print(x)
```

##สร้าง Iterator

การสร้าง object / class ในการวนซ้ำ เราต้องใช้method __iter__() และ __next__()

จากเนื้อหาในเรื่อง classes / Objects class ทุกๆ class มีฟังก์ชั่นที่เรียกว่า __init__() ที่จะใช้เริ่มเมื่อมีการสร้าง class

เราสามารถใช้ __iter__() ได้เช่นเดียวกัน แต่จำเป็นต้องส่งคืนค่าวนซ้ำเอง

__next__() เป็น method ที่จะส่งคืนค่าลำดับถัดไป



##ตัวอย่าง
สร้างการวนซ้ำและส่งคืนค่าตัวเลข โดยเริ่มจาก 1 และเพิ่มขึ้นทีละ 1 ( เช่น ส่งคืนค่า 1 ,2 ,3 ,4 ,5 ):



```python
class MyNumbers:
  def __iter__(self):
    self.a = 1
    return self

  def __next__(self):
    x = self.a
    self.a += 1
    return x

myclass = MyNumbers()
mynum = iter(myclass)

print(next(mynum))
print(next(mynum))
print(next(mynum))
print(next(mynum))
print(next(mynum))
print(next(mynum))
```

##StopIteration

จากตัวอย่างข้างต้นจะเป็นการวนซ้ำไปเรื่อยๆ โดยการใส่ next() ไปเรื่อยๆหรือใช้ for loop

วิธีการป้องกันการวนซ้ำไม่มีสิ้นสุดคือ เราสามารถใช้ StopIteration

เราสามารถป้องกันการ error ของ method __next__() ได้โดยการเพิ่มเงื่อนไขให้สิ้นสุดการวนซ้ำเมื่อครบจำนวนการวนซ้ำที่ต้องการ:

##ตัวอย่าง

หยุดการวนซ้ำเมื่อครบ 15 ครั้ง :



```python
class MyNumbers:
  def __iter__(self):
    self.a = 1
    return self

  def __next__(self):
    if self.a <= 15:
      x = self.a
      self.a += 1
      return x
    else:
      raise StopIteration

myclass = MyNumbers()
myiter = iter(myclass)

for x in myiter:
  print(x)
```
