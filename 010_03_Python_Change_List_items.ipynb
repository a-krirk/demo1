# Python - การเปลี่ยน Item ใน List

## การเปลี่ยนค่า Item
เปลี่ยนค่าของ item ที่ต้องการได้ โดยใช้เลข index อ้างอิง

### ตัวอย่าง
เปลี่ยนค่าของ item ที่ 2 (จาก "papaya" เป็น "orange")

```python
fruits_list = ["apple", "papaya", "banana"]
fruits_list[1] = "orange"
print(fruits_list)
```

## เปลี่ยนค่า Items ใน Range 
เปลี่ยนค่าของ item ต่าง ๆ ใน range ได้โดยการสร้าง list ของ item ใหม่
และอ้างอิง range เลข index ของ item ที่ต้องการเปลี่ยนค่าใหม่

### ตัวอย่าง
เปลี่ยนค่าของ "papaya" และ "banana" เป็น "cherry" และ "kiwi"

```python
fruits_list = ["apple", "papaya", "banana", "orange", "mango"]
fruits_list[1:3] = ["cherry", "kiwi"]
print(fruits_list)
```

ถ้าใส่จำนวน item ใหม่เกินจำนวนที่กำหนดไว้ (ด้วยเลข index)

item ใหม่จะเข้าไปแทนในตำแหน่งที่ระบุไว้ แต่ item เดิมใน index นั้นจะโดนลบทิ้งไปตามลำดับ

### ตัวอย่าง
เปลี่ยนค่าของ item ที่สอง โดยการแทนที่ item ใหม่ไป 2 ตัว

Note: ความยาวของ list จะเปลี่ยนไป เมื่อจำนวนของ item ที่เพิ่มเข้ามาไม่เท่ากับจำนวน item ที่จะโดนแทนที่

```python
fruits_list = ["apple", "papaya", "banana"]
fruits_list[1:2] = ["cherry", "kiwi"]
print(fruits_list)
```

ถ้าใส่จำนวน item ใหม่**น้อยกว่า**จำนวนที่กำหนดไว้ (ด้วยเลข index)

item ใหม่จะเข้าไปแทนในตำแหน่งที่ระบุไว้ แต่ item ที่เหลือ (ตำแหน่งหลัง item ใหม่) ใน list จะโดนลบทิ้งไปทั้งหมด

### ตัวอย่าง
เปลี่ยนค่าของ item ที่สองและสาม โดยแทนที่ด้วย item ใหม่เพียง 1 ตัว

```python
fruits_list = ["apple", "papaya", "banana"]
fruits_list[1:3] = ["cherry"]
print(fruits_list)
```
