#Python ทูเพิล (Tuple)

##ทูเพิล (Tuple)

Tuples ใช้เพื่อเก็บหลายรายการในตัวแปรเดียว

Tuple เป็นหนึ่งใน 4 ประเภทข้อมูลในตัว(built-in)ใน Python ที่ใช้เก็บคอลเลกชั่นข้อมูล อีก 3 ประเภทคือ List, Set และ Dictionary ซึ่งทั้งหมดนี้มีคุณสมบัติและการใช้งานที่แตกต่างกัน

Tuples คือคอลเลกชั่นที่สั่งและไม่สามารถเปลี่ยนแปลงได้

Tuples เขียนด้วยวงเล็บกลม

```python
thistuple = ("mango", "pineapple", "orange")
print(thistuple)
```

##รายการทูเพิล(Tuple Items)

รายการทูเพิลมีการเรียงลำดับ(ordered), ไม่สามารถเปลี่ยนแปลงได้(unchangeable), และอนุญาตให้มีค่าที่ซ้ำกัน(allow duplicate values)

รายการทูเพิลได้รับการจัดทำดัชนี รายการแรกมีดัชนี [0] รายการที่สองมีดัชนี [1] เป็นต้น



##การเรียงลำดับ(Ordered)

เมื่อเราบอกว่ามีการเรียงลำดับtuples หมายความว่ารายการนั้นมีลำดับที่กำหนดไว้และจะไม่เปลี่ยนแปลง



##ไม่สามารถเปลี่ยนแปลงได้(Unchangeable)

ทูเปิลไม่สามารถเปลี่ยนแปลงได้ หมายความว่าเราไม่สามารถเปลี่ยนแปลง เพิ่มหรือลบรายการหลังจากสร้างทูเพิลแล้ว



##อนุญาตให้มีค่าที่ซ้ำกัน(Allow Duplicates)

เนื่องจากทูเพิลถูกทำให้เป็นดัชนี จึงสามารถมีไอเท็มที่มีค่าเท่ากันได้

```python
thistuple = ("mango", "banana", "coconut", "mango", "mango")
print(thistuple)
```

##ความยาวของทูเพิล(Tuple Lenght)

เพื่อพิจารณาว่าทูเพิลมีกี่รายการ ให้ใช้ฟังก์ชัน len()

```python
thistuple = ("apple", "banana", "mango")
print(len(thistuple))
```

##การสร้างทุเพิลด้วยหนี่งรายการ(Create Tuple With One Item)

ในการสร้างทูเพิลที่มีเพียงรายการเดียว คุณต้องเพิ่มเครื่องหมายจุลภาคหลังรายการ มิฉะนั้น Python จะไม่รู้จักว่าเป็นทูเปิล

```python
thistuple = ("apple",)
print(type(thistuple))

#NOT a tuple
thistuple = ("apple")
print(type(thistuple))
```

##รายการทูเพิล -ประเภทข้อมูล

รายการทูเพิลสามารถเป็นข้อมูลประเภทใดก็ได้

```python
tuple1 = ("apple", "mango", "coconut")
tuple2 = (1, 5, 7, 9, 3)
tuple3 = (True, False, False)
```

tuple สามารถประกอบด้วยประเภทข้อมูลที่แตกต่างกันได้

```python
tuple1 = ("สวัสดีวันจันทร์", 60, True, 80, "gretting")
```

##ประเภท type()

tuples ถูกกำหนดให้เป็นวัตถุที่มีประเภทข้อมูล 'tuple'จากมุมมองของPython

```python
mytuple = ("mango", "banana", "coconut")
print(type(mytuple))
```

##ตัวสร้างทูเพิล() [tuple() Constructor]

สามารถใช้ตัวสร้าง tuple() เพื่อสร้าง tuple

```python
thistuple = tuple(("apple", "coconut", "mango")) # สังเกตุวงเล็บ
print(thistuple)
```

##คอลเลกชัน Python (อาร์เรย์) [Python Collections] (Arrays)]

ในภาษาการเขียนโปรแกรม Python มีคอเล็กชั่นของข้อมูล 4 ประเภท
* รายการ(List)เป็นคอลเลกชันที่สั่งและเปลี่ยนแปลงได้ อนุญาตให้สมาชิกที่ซ้ำกัน
* ทูเพิล(Tuple)คือคอลเล็กชันที่สั่งซื้อและไม่สามารถเปลี่ยนแปลงได้ อนุญาตให้สมาชิกที่ซ้ำกัน
* เซ็ท(set)คือคอลเล็กชันที่ไม่มีลำดับและไม่ได้จัดทำดัชนี ไม่มีสมาชิกซ้ำ
* พจนานุกรม(dictionary)คือคอลเล็กชันที่เรียงลำดับได้* และเปลี่ยนแปลงได้ ไม่มีสมาชิกซ้ำ
