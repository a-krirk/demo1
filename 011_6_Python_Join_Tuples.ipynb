# Python - การเชื่อม Tuples

## การเชื่อม 2 Tuples
เราสามารถเชื่อม tuple เข้าด้วยกันโดยใช้เครื่องหมาย + :


## ตัวอย่าง
เชื่อม 2 tuple:

```python
tuple1 = ("B", "O" , "T")
tuple2 = ("N", "O", "I")

tuple3 = tuple1 + tuple2
print(tuple3)
```

## การคูณ Tuples
ถ้าหากต้องการเพิ่มเนื้อหาของ tuple ตามจำนวนครั้งที่่กำหนด สามารถใช้เครื่องหมาย *:


## ตัวอย่าง
เพิ่มจำนวนเนื้อหาของ food tuple 3 ครั้ง:


```python
food = ("Sushi", "Burger", "Salad")
mytuple = food * 3

print(mytuple)
```
