#Python User Input

##User Input

ผู้ใช้งานสามารถใส่ input ไปใน Python ได้

หมายความว่าสามารถถามผู้ใช้งานสำหรับ input ได้

วิธีการมีความแตกต่างกันเล็กน้อยระหว่าง Python 3.6 และ Python 2.7

Python 3.6 ใช้ method **input()**

Python 2.7 ใช้ method **raw_input()**

ตัวอย่างด้านล่างต่อไปนี้ เป็นการถาม username และเมื่อคุณป้อน username เข้าไปก็จะได้ผลลัพธ์ออกมา

##Python 3.6

```python
username = input("Enter username:")
print("Username is: " + username)
```

##Python 2.7

```python
username = raw_input("Enter username:")
print("Username is: " + username)
```

Python จะหยุดประมวลผลจนถึง function **input()** และจะทำต่อเมื่อมีการใส่ค่าไปใน function
