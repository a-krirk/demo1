# Python - Loop Tuples

##การวน Loop Tuple
เราสามารถวน loop tuple ดดยการใช้ for loop


## ตัวอย่าง
วนซ้ำและพิมพ์ค่า :


```python
foodtuple = ("Sushi", "Burger", "Salad")
for x in foodtuple:
  print(x)
```

สามารถเรียนรู้เพิ่มเติมเกี่ยวกับ for loop ได้ที่บทเนื้อหา python for loop ของเรา


## วนซ้ำเลข Index

เราสามารถวนซ้ำ tuple โดยการอ้างอิงถึงเลข Index

เราจะใช้คำสั่ง range() และ len() ในการสร้างการวนซ้ำตามที่ต้องการ


## ตัวอย่าง
แสดงค่าทั้งหมดโดยใช้การอ้างอิงถึงเลข index :

```python
foodtuple = ("Sushi", "Burger", "Salad")
for i in range(len(foodtuple)):
  print(foodtuple[i])
```

##การใช้ While loop

เราสามารถวนซ้ำ list ได้โดยการใช้ while loop

ใช้คำสั่ง len() ในการหาความยาวของ tuple จากนั้นเราจะเริ่มจาก 0 และวนซ้ำไปเรื่อยๆโดยการอ้างอิงเลข index

อย่าลืมที่จะเพิ่มเลข index ขึ้นรอบละ 1



##ตัวอย่าง

แสดงค่าทั้งหมด โดยใช้ while loop :

```python
foodtuple = ("Sushi", "Burger", "Salad")
i = 0
while i < len(foodtuple):
  print(foodtuple[i])
  i = i + 1
```
