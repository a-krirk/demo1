##พจนานุกรม (Dictionary)
พจนานุกรมใช้เพื่อจัดเก็บค่าข้อมูลในรูปของ "คีย์(key):คู่ของคีย์(value pairs)"

พจนานุกรม(Dictionary)คือคอลเล็กชันที่เรียงลำดับได้ เปลี่ยนแปลงได้และไม่อนุญาตให้ทำซ้ำ

Dictionaries are written with curly brackets, and have keys and values:
พจนานุกรม(Dictionary)เขียนด้วยวงเล็บปีกกา และมีคีย์(keys)และค่า(values)

## ตัวอย่าง

## รายการพจนานุกรม [Dictionary Items]

รายการพจนานุกรมมีการเรียงลำดับ(ordered), เปลี่ยนแปลงได้(changeable) และไม่อนุญาตให้มีค่าซ้ำ(Duplicate not allow)

รายการพจนานุกรมจะแสดงเป็น คีย์(keys):คู่คีย์(value pairs)และสามารถอ้างอิงได้โดยใช้ชื่อคีย์(keys)


## ตัวย่าง

```python
thisdict = {
  "brand": "Toyota",
  "model": "Vios",
  "year": 2017}
print(thisdict["brand"])
```

## เรียงลำดับ หรือ ไม่เรียงลำดับ Ordered or Unordered? 

เมื่อเราพูดว่ามีการเรียงลำดับ(Ordered)พจนานุกรม หมายความว่ารายการนั้นมีลำดับที่กำหนดไว้ และลำดับนั้นจะไม่เปลี่ยนแปลง

ไม่เรียงลำดับ(Unordered)หมายความว่าสินค้าไม่มีลำดับที่กำหนดไว้ คุณไม่สามารถอ้างอิงถึงสินค้าโดยใช้ดัชนีได้



## เปลี่ยนแปลงได้ Changeable
พจนานุกรมสามารถเปลี่ยนแปลงได้ หมายความว่าเราสามารถเปลี่ยนแปลง เพิ่ม หรือลบรายการหลังจากที่สร้างพจนานุกรมแล้ว 

## ไม่อนุญาตให้มีค่าซ้ำ(Duplicate not allow)

##คอเล็กชั่นPython(อาเรย์) [Python Collections (Arrays)]

* พจนานุกรม(dictionary)คือคอลเล็กชันที่เรียงลำดับได้ เปลี่ยนแปลงได้ ไม่มีสมาชิกซ้ำ

## การเข้าถึงรายการ (Accessing Items)
สามารถเข้าถึงรายการต่างๆ ของพจนานุกรมโดยอ้างอิงชื่อคีย์ภายในวงเล็บเหลี่ยม:



##ตัวอย่าง

รับค่าของคีย์ "model"



```python
thisdict = {
  "brand": "Toyota",
  "model": "Vios",
  "year": 2017
}
x = thisdict["model"]
```


นอกจากนี้ยังมีวิธีการที่เรียกว่า **get(**) ซึ่งจะให้ผลลัพธ์เช่นเดียวกัน:



## ตัวอย่าง
รับค่าคีย์"model"



```python
x = thisdict.get("model")
```

##รับค่าคีย์(Keys)
keys()จะส่งคืนรายการคีย์ทั้งหมดในพจนานุกรม



## ตัวอย่าง
รับรายการคีย์[list of keys]





```python
x = thisdict.keys()
```

รายการคีย์(List of Keys)เป็นมุมมองของพจนานุกรม หมายความว่าการเปลี่ยนแปลงใดๆ ที่ทำกับพจนานุกรม(dictionary)จะมีผลในรายการคีย์(keys list)



## ตัวอย่าง
เพิ่มรายการใหม่ในพจนานุกรมต้นฉบับ และดูว่ารายการคีย์(keys list)ได้รับการอัปเดตด้วย


```python
car = {
"brand": "Toyota",
"model": "Vios",
"year": 2017
}

x = car.keys()

print(x) #ก่อนเปลี่ยน
car["color"] = "white"

print(x) #หลังเปลี่ยน
```

## รับค่า(Get values)
วิธีการ **values()** จะส่งคืนรายการค่าทั้งหมดในพจนานุกรม



##ตัวยอย่าง
รับรายการของค่า[Get a list of the values]



```python
x = thisdict.values()
รายการค่าต่างๆ เป็นมุมมองของพจนานุกรม หมายความว่าการเปลี่ยนแปลงใดๆ ที่ทำกับพจนานุกรมจะมีผลในรายการค่า
```

รายการ(list)ค่าต่างๆ เป็นมุมมองของพจนานุกรม หมายความว่าการเปลี่ยนแปลงใดๆ ที่ทำกับพจนานุกรมจะมีผลในรายการค่า(values list)


## ต้วอย่าง
ทำการเปลี่ยนแปลงในพจนานุกรมต้นฉบับ และดูว่ารายการค่าได้รับการอัปเดตด้วย:



```python
car = {
"brand": "Toyota",
"model": "Vios",
"year": 2017
}

x = car.values()

print(x) #ก่อนเปลี่ยน
car["year"] = 2020

print(x) #หลังเปลี่ยน
```

## ตัวอย่าง
เพิ่มรายการใหม่(new items)ในพจนานุกรมต้นฉบับ และดูว่ารายการค่า(values list)ได้รับการอัปเดตด้วย


```python
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

x = car.values()

print(x) #ก่อนเปลี่ยน

car["color"] = "red"

print(x) #หลังเปลี่ยน
```

#รับรายการ [Get Items]
items() จะส่งคืนแต่ละรายการในพจนานุกรม เป็นtuples ในรายการ



## ตัวอย่าง
รับรายการ คีย์(key):คู่คีย์(value pairs)




```python
x = thisdict.items()
```

รายการ(list)ที่ส่งคืนคือมุมมองของรายการในพจนานุกรม หมายความว่าการเปลี่ยนแปลงใดๆ ที่ทำกับพจนานุกรมจะมีผลในรายการ(items list)

##ตัวอย่าง
ทำการเปลี่ยนแปลงในพจนานุกรมต้นฉบับ และดูว่ารายการ(items list)ได้รับการอัปเดตด้วย



```python
car = {
"brand": "Toyota",
"model": "Vios",
"year": 2017
}

x = car.items()

print(x) #ก่อนเปลี่ยน

car["year"] = 2020

print(x) #หลังเปลี่ยน
```

## ตัวอย่าง
เพิ่มรายการใหม่(new item)ในพจนานุกรมต้นฉบับ และดูว่ารายการได้รับการอัปเดตด้วย:



```python
car = {
"brand": "Toyota",
"model": "Vios",
"year": 2017
}

x = car.items()

print(x) #ก่อนเปลี่ยน

car["color"] = "red"

print(x) #หลังเปลี่ยน
```

## ตรวจสอบคีย์(Keys)ที่มี
เพื่อพิจารณาว่ามีคีย์ที่ระบุอยู่ในพจนานุกรมหรือไม่ ให้ใช้คีย์เวิร์ด **in**



##ตัวอย่าง
ตรวจสอบว่า "model" มีอยู่ในพจนานุกรมหรือไม่

```python
thisdict = {
  "brand": "Toyota",
  "model": "Vios",
  "year": 2017
}
if "model" in thisdict:
  print("Yes, 'model' is one of the keys in the thisdict dictionary")
```

# Python - การเปลี่ยนรายการพจนานุกรม

## อัพเดทพจนานุกรม
ใช้ update()จะอัปเดตพจนานุกรมด้วยรายการจากอาร์กิวเมนต์ที่กำหนด

อาร์กิวเมนต์ต้องเป็นพจนานุกรม หรืออ็อบเจ็กต์แบบ iterable ที่มี คีย์(key):คู่คีย์(value pairs)


##ตัวอย่าง

## เพิ่มรายการ (Adding Items)

## Python - ลบรายการพจนานุกรม


## ลบรายการ(Removing Items)
มีหลายวิธีที่สามารถลบรายการจากพจนานุกรมได้




## ต้วอย่าง

วิธีการ **pop()** จะลบรายการที่มีชื่อคีย์(key name)ที่ระบุ:



```python
thisdict = {
  "brand": "Toyota",
  "model": "Vios",
  "year": 2017
}
thisdict.pop("model")
print(thisdict)
```

##ตัวอย่าง
วิธี **popitem()** จะลบรายการที่แทรกล่าสุด (ในเวอร์ชันก่อน 3.7 รายการสุ่มจะถูกลบออกแทน):

```python
thisdict = {
  "brand": "Toyota",
  "model": "Vios",
  "year": 2017
}
thisdict.popitem()
print(thisdict)
```

##ตัวอย่าง
คีย์เวิร์ด **del** ลบรายการที่มีชื่อคีย์ที่ระบุ:



```python
thisdict = {
  "brand": "Toyota",
  "model": "Vios",
  "year": 2017
}
del thisdict["model"]
print(thisdict)
```

## ตัวอย่าง
 **del** สามารถลบพจนานุกรมทั้งหมดได้:




```python
thisdict = {
  "brand": "Toyota",
  "model": "Vios",
  "year": 2017
}
del thisdict
print(thisdict) #จะทำให้เกิดerrorเพราะ "thisdict" ไม่มีอยู่อีกต่อไป
```

## ตัวอย่าง
**clear()** จะล้างพจนานุกรม:



```python
thisdict = {
  "brand": "Toyota",
  "model": "Vios",
  "year": 2017
}
thisdict.clear()
print(thisdict)
```

## Python - การวนลูป(loop)พจนานุกรม


## การวนลูป(loop)ผ่านพจนานุกรม
สามารถวนลูปพจนานุกรมโดยใช้ for loop

เมื่อวนลูปซ้ำผ่านพจนานุกรม ค่าที่ส่งคืน(return value)คือคีย์ของพจนานุกรม(keys of the dictionary)แต่มีวิธีการอื่นที่ส่งคืนค่าคู่คีย์(values)ด้วย

## ตัวอย่าง
พิมพ์ชื่อคีย์(key names)ทั้งหมดในพจนานุกรม ทีละชื่อ:



```python
for x in thisdict:
  print(x)
```

## ตัวอย่าง

```python
for x in thisdict:
  print(thisdict[x])
```

## ตัวอย่าง
สามารถใช้วิธี **values()** เพื่อคืนค่าของพจนานุกรม:



```python
for x in thisdict.values():
  print(x)
```

## ตัวอย่าง
สามารถใช้วิธี **values()** เพื่อคืนค่าของพจนานุกรม:



```python
for x in thisdict.keys():
  print(x)
```

## ตัวอย่าง
วนลูปซ้ำทั้งคีย์และค่าโดยใช้ items() วิธีการ:




```python
for x, y in thisdict.items():
  print(x, y)
```

##Python - คัดลอกพจนานุกรม

## คัดลอกพจนานุกรม
คุณไม่สามารถคัดลอกพจนานุกรมเพียงแค่พิมพ์ **dict2 = dict1** เนื่องจาก: **dict2** จะเป็นการอ้างอิงถึง **dict1** และการเปลี่ยนแปลงที่ทำใน **dict1** จะทำโดยอัตโนมัติใน **dict2**.

มีวิธีการทำสำเนา วิธีหนึ่งคือใช้วิธีพจนานุกรมในตัว

## ตัวอย่าง
คัดลอกพจนานุกรมด้วยวิธี **copy()**:



```python
thisdict = {
  "brand": "Toyota",
  "model": "Vios",
  "year": 2017
}
mydict = thisdict.copy()
print(mydict)
```

อีกวิธีหนึ่งในการคัดลอกคือการใช้ฟังก์ชัน dict() ในตัว



## ตัวอย่าง
ทำสำเนาพจนานุกรมด้วยฟังก์ชัน dict():


```python
thisdict = {
  "brand": "Toyota",
  "model": "Vios",
  "year": 2017
}
mydict = dict(thisdict)
print(mydict)
```
