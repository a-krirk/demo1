#Python Modules

##อะไรคือ Module ?

module คล้ายกับ library คือ เป็นไฟล์ที่รวบรวมฟังก์ชั่นต่างๆของเราเอาไว้




##การสร้าง Module

การสร้าง module เพื่อบันทึก code ที่เราต้องการ ไฟล์นั้นจะต้องมีนามสกุลเป็น .py


บันทึก code นี้ในไฟล์ที่ชื่อ mymodule.py


```python
def Food(menu):
  print("I'd like " + menu)
```

##การใช้ Module

ตอนนี้เราสามารถใช้ module ที่สร้างไปข้างต้นโดยการใช้คำสั่ง import :




```python
import mymodule

mymodule.Food("Sushi")
```

##ตัวแปรใน Module

ภายใน module ก็จะมีฟังก์ชั่นตามที่อธิบายไว้ข้างต้น และยังรวมถึงตัวแปลทุกประเภท (arrays, dictionaries, objects ฯลฯ):



บันทึกไฟล์ code นี้ชื่อว่า mymodule.py


```python
menu1 = {
  "menu": "Sushi",
  "price": 36,
  "nationality": "Japanese"
}
```

นำเข้า module ชื่อว่า mymodule และเข้าถึง dictionary menu1 :


```python
import mymodule

a = mymodule.menu1["price"]
print(a)
```

##การตั้งชื่อ Module

เราสามารถตั้งชื่อ module ได้ตามที่ต้องการ แต่ไฟล์นั้นจำเป็นที่จะต้องมีนามสกุล .py




##การเปลี่ยนชื่อ Module

เราสามารถสร้างนามแฝงให้ module ได้ตอนเรา import module โดยการใช้คำสั่ง as :


สร้างนามแฝงให้กับ mymodule โดยตั้งชื่อว่า bn :


```python
import mymodule as bn

a = bn.menu1["price"]
print(a)
```

##Built-in Modules

มี built-in modules หลายตัวใน python ที่เราสามารถ import มาใช้งานได้ตามต้องการ



นำเข้าและใช้ module platform:


```python
import platform

A = platform.system()
print(A)
```

##การใช้ฟังก์ชั่น dir()

คำสั่ง dir() เป็นคำสั่งที่ใช้สำหรับแสดงรายการชื่อฟังก์ชั่นทั้งหมด (หรือชื่อตัวแปร) ในmodule


แสดงรายชื่อฟังกืชั่นทั้งหมดของ platform module:


```python
import platform

A = dir(platform)
print(A)
```

##นำเข้าจาก Module

เราสามารถเลือกนำเข้าบางส่วนของ module โดยการใช้คำสั่ง from


module mymodule มี 1 ฟังก์ชั่น และ 1 dictionary:


```python
def Food(menu):
  print("I'd like " + menu)

menu1 = {
  "menu": "Sushi",
  "price": 36,
  "nationality": "Japanese"
}
```

นำเข้าเฉพาะ dictionary menu1 จาก module


```python
from mymodule import menu1

print (menu1["price"])
```
