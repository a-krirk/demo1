# Python String Formatting

เพื่อให้มั่นใจว่า string จะแสดงผลลัพธ์ตามที่ต้องการ เราสามารถจัดรูปแบบผลลัพธ์ได้ด้วย method **format()**

## String format()

สามารถจัดรูปแบบส่วนต่าง ๆ ของ  string ได้ด้วย method **format()** 

บางครั้งมีบางส่วนของข้อความที่ไม่ต้องการ ซึ่งอาจจะมาจากใน database หรือ input ที่รับเข้ามา

สามารถปรับรูปแบบของข้อความได้ด้วยการใส่ **curly brackets {}** ลงในไปข้อความนั้น และรันค่าผ่าน method **format()**

## ตัวอย่าง
ใส่ **curly brackets {}** ลงในไปส่วนที่ต้องการแสดงค่า price

```python
price = 36
txt = "The price is {} dollars"
print(txt.format(price))
```

สามารถเพิ่ม parameter ภายใน **curly brackets** เพื่อระบุการปรับเปลี่ยนรูปแบบของข้อมูล

## ตัวอย่าง
จัดรูปแบบ price ให้แสดงผลเป็นทศนิยม 2 ตำแหน่ง

```python
txt = "The price is {:.2f} dollars"
```

##Multiple Values

ถ้าต้องการใช้ค่าหลาย ๆ ค่า ทำได้ด้วยการเพิ่มค่าเหล่านั้นไปใน method **format()**

## ตัวอย่าง
ใส่จำนวน **curly brackets {}** ไปมากขึ้น (ตามจำนวนข้อมูลที่ใส่เข้าไป)

```python
quantity = 3
itemno = 567
price = 49
myorder = "I want {} pieces of item number {} for {:.2f} dollars."
print(myorder.format(quantity, itemno, price))
```

## เลข Index

สามารถเพิ่มเลข index (ตัวเลขที่อยู่ใน curly brackets เช่น {0})
เพื่อมั่นใจได้ว่าข้อมูลจะถูกจัดวางไว้ในตำแหน่งที่ถูกต้อง

```python
quantity = 3
itemno = 567
price = 49
myorder = "I want {0} pieces of item number {1} for {2:.2f} dollars."
print(myorder.format(quantity, itemno, price))
```

ใช้เลข **index** ถ้าคุณต้องการอ้างถึงข้อมูลเดียวกันมากกว่า 1 ครั้ง

```python
age = 29
name = "Botnoi"
txt = "His name is {1}. {1} is {0} years old."
print(txt.format(age, name))
```

## ชื่อ Index

สามารถใช้**ชื่อ index** ได้ด้วยการใส่ชื่อไปใน **curly bracket** เช่น {carname} หลังจากนั้นจะต้องใช้ชื่อ index เมื่อใส่ค่า parameter เข้าไป เช่น txt.format(carname = "Ford")

```python
myorder = "I have a {carname}, it is a {model}."
print(myorder.format(carname = "Ford", model = "Mustang"))
```
