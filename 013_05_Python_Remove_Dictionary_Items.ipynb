# Python - การลบ Items ใน Dictionary 

## การลบ Items 
มีหลายวิธีที่สามารถลบ item ใน dictionary ได้

### ตัวอย่าง
ใช้ method **pop()** ลบ item ด้วยการระบุชื่อ key

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.pop("model")
print(thisdict)
```

### ตัวอย่าง
ใช้ method **popitem()** ลบ item สุดท้ายที่เพิ่มเข้าไปใน dictionary

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.popitem()
print(thisdict)
```

### ตัวอย่าง
ใช้ keyword  **del** ลบ item ด้วยการระบุชื่อ key

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
del thisdict["model"]
print(thisdict)
```

### ตัวอย่าง
ใช้ keyword  **del** ลบ dictionary

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
del thisdict
print(thisdict) #this will cause an error because "thisdict" no longer exists.
```

### ตัวอย่าง
ใช้ method **clear()** เพื่อลบ item ทั้งหมดใน dictionary

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.clear()
print(thisdict)
```
